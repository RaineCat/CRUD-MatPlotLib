```
#dictionary of questions and answers
QandA=[{'question':'Who was zeus’s wife?\nA. Hera\nB. Athena\nC. Aphrodite\nD. Artemis','answer':'a'},
        {'question':'Aphrodite was the goddess of\nA. Love and Beauty\nB. Beauty and War\nC. Love and Hate\nD. Anger and Sex','answer':'a'},
        {'question':'How was Athena born\nA. From Zeus’s head\nB. From the flowers\nC. From Hera’s womb\nD. From the ocean','answer':'a'},
        {'question':'Who was the God of war\nA. Ares\nB. Hephaestus\nC. Hades\nD. Dionesyus','answer':'a'},
        {'question':'Icarus flew too close to the what\nA. Sun\nB. Moon\nC. Sea\nD. Sand','answer':'a'}]

#*************game proper*********************************************
def gameprop(): 
  namelist=[] #list of previous players
  scorelist=[] #list of previous players scores
  endgame="" #to end the game
  while endgame!='quit':
    print("Welcome into the World of Greek Mythology") 
    name=input("Let's start by carving your name onto this tablet: ") #some user inputs
    print("\nWelcome "+ name.title() + "!"+"\n\nMake sure to finish this quiz\nThere's no turning back now!")
    print('Type the letter of your answers\n\n\nPress any key to start')
    enter=input() #for press any key to start, not too important

    score=0 #initial score
    for item in QandA: #calling all questions inside the dictionary
        print(item['question'])
        answer=input('Answer: ')
        if answer.lower()==item['answer']: #checking the answer
            score +=1
            print('\nYass! Way to go!\n')
        else: 
            print('\nBoo!\n')

    #rating() #call out the rating function
    rate=(score/len(QandA))*100
    #showing the result
    print('\n\n************************************')
    print('\nYour score is ', str(score), 'out of 5')
    if rate == 100:
      print('Yaay! Perf!')
    elif rate < 100 and rate >=70:
      print('Fair enough!')
    else:
      print('Yikes!')

    #player_list() #call out the previous players
    namelist.append(name)
    scorelist.append(score)
    print('Score board of captives:\n')

    for i in range(len(namelist)):
      #for scores in scorelist:
      print(namelist[i] +' : ' + str(scorelist[i]) +'\n')

    endgame=input('\n---------------------------\nType "QUIT" to  end the game\n')
    
  print('You escaped the worlds of gods and goddesses!')
#*********************************************************************

#*************calculating the rating**********************************
def rating(): 
  rate=(score/len(QandA))*100
  #showing the result
  print('\n\n************************************')
  print('\nYour score is ', str(score), 'out of 5')
  if rate == 100:
    print('Yaay! Perf!')
  elif rate < 100 and rate >=70:
    print('Fair enough!')
  else:
    print('Yikes!')
#*********************************************************************

#************adding list of names and scores of previous players******
def player_list(): 
  namelist=[] #list of previous players
  scorelist=[] #list of previous players scores
  namelist.append(name)
  scorelist.append(score)
  print('List of captives:\n')
  for i in range(len(namelist)):
    #for scores in scorelist:
    print(namelist[i] +' : ' + str(scorelist[i]) +'\n')
#*********************************************************************

#***************Actual CRUD*******************************************
#delete items
def delQandA():
  item=input('Enter question number:\t')
  item_no=int(item)-1
  print(QandA[item_no]['question'])
  print(QandA.pop(item_no))
  print(QandA)

#add items
def addQandA():
  addQuest=input('Type the question you want to add using this format:\n*QUESTION* nA. *choice* nB. *choice* nC. *choice* nD. *choice*\nAdd back slash before the letter "n"\n')
  addAns=input('Input only the letter of the correct answer:\t')
  QandA.append({'question':addQuest,'answer':addAns.lower()})
  print(QandA)

#replace items
def replQandA():
  item=input('Enter question number:\t')
  item_no=int(item)-1
  print(QandA.pop(item_no))

  addQuest=input('Type the question you want to add using this format:\n*QUESTION* nA. *choice* nB. *choice* nC. *choice* nD. *choice*\nAdd back slash before the letter "n"\n')
  addAns=input('Input only the letter of the correct answer:\t')
  QandA.insert(item_no, {'question':addQuest,'answer':addAns.lower()})
  print(QandA)

#*********************************************************************

#***************intro to CRUD*****************************************
def CRUD():
  print('Do you want to update the question bank?\n')
  editmoghorl=input('Yes/No\n')
  if editmoghorl.lower()=='no':
    playermode()
  elif editmoghorl.lower()=='yes':
    print('Do you want to:\n')
  editIt=int(input('1- Delete question\n2- Replace question\n3- Add question\n4- Exit\n'))
  while editIt<1 or editIt>3: #input validation
    print('Invalid input')
    editIt=int(input('1- Delete question\n2- Replace question\n3- Add question\n4- Exit\n'))
    continue
  if str(editIt)=='1':
    delQandA()
  elif str(editIt)=='2':
    replQandA()
  elif str(editIt)=='3':
    addQandA()
  else:
    playermode()

#*********************************************************************

#***************player or moderator plus validation*******************
def playermode():
  validation=False
  while validation == False:
    Player_gm=input('Are you a Player or a Moderator?\nType\n0 - Player\n1 - Moderator\t  ')
    try:
      Player_gm=int(Player_gm)
      if Player_gm < 0 or Player_gm > 1:
        print('Invalid input')
        continue
      else:
        validation=True
    except:
      print('Invalid input')

  if Player_gm==0:
    gameprop()
  else:
    CRUD()
#*********************************************************************
playermode()

```
